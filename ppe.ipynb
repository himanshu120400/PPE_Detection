```python
import torch

print(torch.cuda.is_available())
```

```python
import os
from ultralytics import YOLO

DATA_YAML_PATH = 'ppe_11_classes_rf_himanshu_dataset/data.yaml'

model = YOLO('yolov8m.pt')

results = model.train(
    data=DATA_YAML_PATH,  
    epochs=10,           
    imgsz=640,            
    batch=8,             
    conf=0.25,            
    iou=0.4,  
    device='cpu',            
    name='ppe_compliance_train'
)

print("\nTraining completed. Results saved in runs/detect/ppe_train")

RUN_DIR = 'runs/detect/ppe_compliance_train'
```

```python
import matplotlib.pyplot as plt

try:
    results_image_path = os.path.join(RUN_DIR, 'results.png') 
    
    if os.path.exists(results_image_path):
        img = plt.imread(results_image_path)
        plt.figure(figsize=(15, 10))
        plt.imshow(img)
        plt.title('Validation and Training Loss/Metrics')
        plt.axis('off')
        plt.show()
    else:
        print("Error: results.png not found. Check the RUN_DIR path.")

except Exception as e:
    print(f"An error occurred while displaying graphs: {e}")
```
